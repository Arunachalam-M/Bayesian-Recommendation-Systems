```python
import pandas as pd
import numpy as np
import random
from sklearn import preprocessing
```

```python
movie_details = pd.read_csv('data/movies_metadata.csv')
```

```python
ratings = pd.read_csv('data/ratings_small.csv')
```

```python
predicted = pd.read_csv('ratings2.csv')
```

```python
movie_details = movie_details[['id','title','budget','popularity','vote_average', 'vote_count']]
```

```python
ratings = ratings[['userId', 'movieId', 'rating']]
```

```python
users = np.array(ratings['userId'])
users = np.unique(users)
movies = np.array(ratings['movieId'])
movies = np.unique(movies)
ratings_ij = np.array(ratings['rating'])
```

```python
m = len(users)
print("Users:", m)

n = len(movies)
print("Movies:", n)
```

```python
movie_map = {}
reverse_map = {}

for i in range(n):
    movie_map[movies[i]] = i+1
    reverse_map[i] = movies[i]
```

```python
ratings['movieId'] = ratings.apply(lambda x: movie_map[x['movieId']], axis = 1)
```

```python
movie_details = movie_details[movie_details['id'].str.isnumeric()]
```

```python
movie_details.head()
```

```python
movie_details['id'] = pd.to_numeric(movie_details['id'])
```

```python
movie_details['movieId'] = movie_details.apply(lambda x: movie_map[x['id']] if x['id'] in movie_map.keys() else 0, axis = 1)
```

```python
movie_details[movie_details['id'] <100]
```

```python
ratings.head()
```

```python
res1 = ratings.merge(movie_details[['movieId','title','vote_average']], on=['movieId'], how='inner')
```

```python
res1 = res1.sort_values(by=['userId'])
```

```python
movie_names = {}
for i in range(len(res1)) : 
    movie_names[res1.iloc[i]['movieId']] = res1.iloc[i]['title']
```

```python
res1[res1['userId']==5]
```

```python
predicted.columns
```

```python
predicted.drop(['Unnamed: 0'], axis=1, inplace = True)
```

```python
predicted = preprocessing.scale(predicted, axis=1)
```

```python
predicted = predicted + 2.5
```

```python
predicted.shape
```

```python
def recommendations(user_id):
    preds = np.array(predicted[user_id-1,])
    recoms = preds.argsort()[-50:][::-1]
    recoms = recoms + 1
    titles = [movie_names[i] for i in recoms if i in movie_names]
    return titles[:10]
```

```python
recommendations(1)
```

```python
def rated(user_id):
    df = res1[res1['userId']==user_id]
    df = df.sort_values(by=['rating'], ascending = False)
    return df['title'].to_numpy().tolist()[:10]
```

```python
rated(1)
```

```python
user = random.randint(1,600)
print (user)
print("Seen :", rated(user))
print("recommended :", recommendations(user))
```

```python
ratings['predicted'] = ratings.apply(lambda x: predicted[int(x['userId'])-1,int(x['movieId'])-1], axis = 1)
```

```python
ratings.to_csv('comparison_normalized.csv')
```
